```
logs= sc.textFile("/FileStore/logs")
logs.take(2)
```

```
userreqRDD=logs.map(lambda line: line.split()).map(lambda s: (s[2], 1)).reduceByKey(lambda a,b: a+b)
userreqRDD.take(4)
```

```
accounthits= KeyRDD.join(userreqRDD).values()
accounthits.take(2)
```

```
for line in accounthits.toDebugString().decode().splitlines():
    print(line)
```
